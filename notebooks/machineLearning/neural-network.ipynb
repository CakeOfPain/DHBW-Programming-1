## Aktivierungsfunktion Sigmoid

```python
import math
import pandas

def sigmoid(x):
    return 1 / (1 + math.exp(x * -1))

values = [sigmoid(x / 10.0) for x in range(-100, 100, 1)]

df = pandas.DataFrame({'sigmoid': values})
df.plot(title="DataFrame Plot")
```
